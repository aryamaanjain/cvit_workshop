## 3. Clean data

```
# Task 3.1
df.columns = ['track', 'artist', 'genre', 'bpm', 'energy', 'danceability', 'loudness', 'liveness', 'valence', 'length', 'acousticness', 'speechiness', 'popularity']
df.head()
```

```
# Task 3.2.1, dropping rows
df1 = df.copy()
df1.dropna(how='any')
```

```
# Task 3.2.2.1, mean
df2 = df.copy()
df2.fillna(df2.mean())
```

```
# Task 3.2.2.1, median
df2 = df.copy()
df2.fillna(df2.median())
```

## 4. Query data

```
artist = df[df['popularity'] == min(df['popularity'])].iloc[0]['artist']
loudest = max(df.loc[df['artist'] == artist]['loudness'])
print(df[(df['artist']==artist) & (df['loudness']==loudest)].iloc[0]['track'])
```
