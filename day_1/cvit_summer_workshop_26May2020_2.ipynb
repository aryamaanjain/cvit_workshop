```
import numpy as np
import matplotlib.pyplot as plt
```

```
# Plot x=25
a = 1
b = 0
c = -25

Y = np.linspace(start=-100, stop=100, num=200)
X = (-b*Y -c) / a

plt.plot(X, Y)
plt.show()
```

```
# Plot y=50
a = 0
b = 1
c = -50

X = np.linspace(start=-100, stop=100, num=200)
Y = (-a*X -c) / b

plt.plot(X, Y)
plt.show()
```

```
# Plot x=20 and w-vector
a = 1
b = 0
c = -20

Y = np.linspace(start=-100, stop=100, num=200)
X = (-b*Y -c) / a

scale_factor = 20
w = scale_factor * np.array([a, b])

plt.plot(X, Y, label = 'line')
plt.plot([0, w[0]], [0, w[1]], label = 'w-vector')
plt.legend(loc = "lower left")
plt.show()
```

```
import math
def distance_from_origin(a, b, c):
  return abs(c) / math.sqrt(a^2 + b^2)
```

```
def distance_parallel_lines(a, b, c1, c2):
  return abs(c2 - c1) / math.sqrt(a^2 + b^2)
```

```
# line x = 50 distance from origin
print(distance_from_origin(1, 0, -50))

# line y = 25 distance from origin
print(distance_from_origin(0, 1, -25))
```

```
# tanh function
x = np.linspace(-10, 10, 200)
y = (np.exp(x) - np.exp(-x)) / (np.exp(x) + np.exp(-x))
plt.plot(x, y)
plt.show()
```

```
# gaussian function
x = np.linspace(-4, 4, 100)
y = (np.exp(-x*x))
plt.plot(x, y)
plt.show()
```

```
# ReLU function
x = np.linspace(-10, 10, 200)
y = np.maximum(x, np.zeros(x.size))
plt.plot(x, y)
plt.show()
```

```
# Leaky ReLU function
x = np.linspace(-10, 10, 200)
y = np.maximum(x, 0.1*x)
plt.plot(x, y)
plt.show()
```

```
# step function
def step(a):
  if a >= 0:
    return 1
  else:
    return 0

x = np.linspace(-5, 5, 100)
y = np.array([step(i) for i in x])
plt.plot(x, y)
plt.show()
```

```
# natural log
x = np.linspace(0.01, 10, 200)
y = np.log(x)
plt.plot(x, y)
plt.show()
```

```
# exponential
x = np.linspace(-5, 5, 200)
y = np.exp(x)
plt.plot(x, y)
plt.show()
```

```
# square
s = 10

x = np.linspace(-s, s, 200)
y1 = [s for i in range(200)]
y2 = [-s for i in range(200)]
plt.plot(x, y1)
plt.plot(x, y2)

y = np.linspace(-s, s, 200)
x1 = [s for i in range(200)]
x2 = [-s for i in range(200)]
plt.plot(x1, y)
plt.plot(x2, y)

plt.show()
```

```
# rectangle
l = 20
b = 10

x = np.linspace(-l, l, 200)
y1 = [b for i in range(200)]
y2 = [-b for i in range(200)]
plt.plot(x, y1)
plt.plot(x, y2)

y = np.linspace(-b, b, 200)
x1 = [l for i in range(200)]
x2 = [-l for i in range(200)]
plt.plot(x1, y)
plt.plot(x2, y)

plt.show()
```

```
# ellipse
x = np.linspace(-2, 2, 200)
y = np.sqrt(1 - (x*x)/4)
plt.plot(x, y)
plt.plot(x, -y)
plt.show()
```

```
x = np.linspace(-2, 2, 200)
y = np.linspace(-2, 2, 200)
X, Y = np.meshgrid(x, y)
f = np.exp(-X*X - Y*Y)
plt.contour(X, Y, f)
plt.show()
```

```
# 3D plot
a = 1
b = 1
c = 0
d = 10

x = np.linspace(-10, 10, 20)
z = np.linspace(-10, 10, 20)
xx, zz = np.meshgrid(x, z)
yy = (-a*xx - c*zz - d) / b

plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(xx, yy, zz)
```

```
# 2D gaussian function
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-2, 2, 50)
y = np.linspace(-2, 2, 50)
xx, yy = np.meshgrid(x, y)
zz = np.exp(-(xx*xx + yy*yy))

plt3d = plt.figure().gca(projection='3d')
plt3d.plot_surface(xx, yy, zz)
```

```
# scatter plot
from sklearn.datasets import make_swiss_roll
X, _ = make_swiss_roll(1000)

xs = X[:, 0]
ys = X[:, 1]
zs = X[:, 2]

from mpl_toolkits.mplot3d import Axes3D
plt3d = plt.figure().gca(projection='3d')
plt3d.scatter(xs, ys, zs)
```
